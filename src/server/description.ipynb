```python
def count_regex_matches(df, patterns_dict):
    """
    Counts rows in df['description'] matching each regex pattern.
    Counts rows in df['title'] matching each regex pattern.
    
    Parameters:
    df (pd.DataFrame): DataFrame with 'description' column
    patterns_dict (dict): {pattern_name: regex_str}
    
    Returns:
    dict: {pattern_name: count}
    """
    for name, regex in patterns_dict.items():
        desc_matches = df['description'].str.contains(regex, regex=True, na=False)
        print(desc_matches)
        titl_matches = df['title'].str.contains(regex, regex=True, na=False)

        descs = desc_matches.sum()
        titls = titl_matches.sum()
        total = (desc_matches | titl_matches).sum()
        print(f"{f'{name:<20}'[:20]} : found {large_number_for_display(descs)} in 'description', {large_number_for_display(titls)} in 'title', TOTAL: {large_number_for_display(total)}")
```

```python
count_regex_matches(photos, {
    'year_dash': r'--\d{4}',
    'year_range': r"\d{4}-\d{4}",
    'year_range_soft': r"\d{4}s?\s?-\s?\d{4}\s?"
})
```

```python
count_regex_matches(photos, {
    'bracket date' : r'\[\d{4}\]',
    'bracket date newline' : r'\n\[\d{4}\]\n',
})
```

```python
count_regex_matches(photos, {
    'date field' : r'Date:',
    'format field' : r'Format:',
})
```

```python
count_regex_matches(photos, {
    'after' : r'after',
    'before' : r'before',
    'circa' : r'circa|Circa',
})
```

```python
count_regex_matches(photos, {
    'decade': r'(?<!\d)\d{4}s',
    'raw year': r'(?<!\d)\d{4}(?!\d)'
})
```

```python
count_regex_matches(photos, {
    'World war 2' : r'WWII(?!I)',
    'World war 1': r'WWI(?!I)'
})
```
